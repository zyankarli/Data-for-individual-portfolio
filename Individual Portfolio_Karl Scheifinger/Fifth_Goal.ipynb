## Introduction
In times of continuous environmental crises, governments and scientists search for new methods to gather data about the natural environment to overcome the pitfalls of reactionary environmental governance. One new approach on the rise in Smart Earth, which can be briefly defined as a ‘set of environmental applications of the Internet of Things’ (Bakker & Ritts, 2018, p.1), ranging from remote sensing techniques to Environmental Sensor Networks, which can be thought of as interconnected sensor arrays (Hart & Martinez, 2006). Smart Earth applications provide enormous amount of environmental data, thereby potentially transforming environmental governance practises.
This report will examine two concrete examples of Smart Earth applications: the Ocean Observatory Initiative and geo-fencing by Safe The Elephants.


## The Ocean Observatory Initiative
The first example is the Ocean Observatory Initiative (OOI). In brief, the OOI can be described as an attempt to “wire the oceans“ (Paul, K., 2015). The OOI currently features five arrays of sensors that together form a sensor complex of more than 800 instruments which are used to generate live ocean data, amongst others measuring water temperature at different sea depths, acidity and oxygen content (OOI, About OOI, n.d.). <p>

While the overall goal of the OOI is to better understand the planets’ oceans and to monitor changes in the deep sea, the program is said to have enormous potential for scientific applications. For example its data can significantly improve current climate models (OOI, The Vision, n.d.). On the other side, also private companies are expected to profit, for example through enhanced early warning systems for maritime fisheries that detect increasing acidity levels due to upwellings (Paul, K., 2015). All data generated within the OOI project is freely accessible on the internet (available at https://dataexplorer.oceanobservatories.org/#go-to-data-access).<br> This brief case study shows that smart environments can bring change to environmental governance by providing great amounts of detailed live data from otherwise inaccessible areas.


```python
#Display current arrays of OOI
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

#Get world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

#Locations of central mooring locations of OOI arrays
Station = ['Global Station Papa Array','Cabeld Array','Coastal Endurance Array','Coastal Pioneer Array','Irminger Sea']
Lat = [50,46,45,40,60]
Long = [-145, -127, -124, -71, -39]

#Create dataframe
df = pd.DataFrame(Station)
df['Long'] = Long
df['Lat'] = Lat
df

#Create geopandas dataframe
gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df.Long, df.Lat))
#Plot arrays
ax = world.plot(color='white', edgecolor='black')
gdf.plot(ax=ax, color='red')
plt.show()
```

## Save The Elephants
The second example is the geofencing-system applied by Save The Elephants (STE). STE is a research and conservation organisation that operates in Northern Kenya with the objective to sustain the local elephant population (STE, About STE, n.d.). In order to reduce negative elephants impacts on human population, i.e. crop raids, in so called conflict-zones, a geofencing system was implemented. Single elephants are equipped with collars that transmit their GPS location in real-time. In case an individuum leaves the reserve area and enters individual property, automatically generated text messages are sent to the mobile phones of local famers which than can drive back the elephant before their crops get damaged (STE, Geofencing, n.d.). Profits through the real time elephant tracking were not only generated for local farmers but also for the global scientific community that was able to study elephant movements in unprecedented detail. <br> 
Due to its success, the project is now meant to be expanded to other conflict zones in Kenya (STE, Geofencing, n.d.). <br>This second example of shows that smart environments can critically enhance preventive measures in environmental governance by providing specific live data to specific actors at the right time.
